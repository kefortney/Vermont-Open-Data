```python
%matplotlib inline

# Handle table-like data and matrices
import numpy as np
import pandas as pd

# Visualisation
import matplotlib 
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab
import seaborn as sns
```

```python
wholedf = pd.read_csv('Income_Stats_2014_Family_By_Town.csv')
```

```python
wholedf.head()
```

```python
wholedf.sort_values(by='Median AGI per Return',ascending=0)
```

```python
wholedf['taxoverincome'] = wholedf['Net Vermont Tax'] / wholedf['Adjusted Gross Income']
wholedf['perctax'] = wholedf['Net Vermont Tax'] / 435880094
wholedf['percreturns'] = wholedf['Returns'] / 149876
wholedf['returnexempt'] = (wholedf['Exempt'] - wholedf['Returns']) / wholedf['Returns']
```
